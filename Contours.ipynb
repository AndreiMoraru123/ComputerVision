```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('thumbs_up_down.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)

plt.show()
```

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGBA2GRAY)

retval, binary = cv2.threshold(gray, 215, 255, cv2.THRESH_BINARY_INV)

plt.imshow(binary,cmap='gray')

plt.show()
```

```python
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

image_copy2 = np.copy(image)

all_cont = cv2.drawContours(image_copy2,contours,-1,(0,255,0),2)

plt.imshow(all_cont)
```

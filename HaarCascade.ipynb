```python
# import required libraries for this section
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('multi_faces.jpg')

image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_RGBA2GRAY)

plt.figure(figsize=(20,10))
plt.imshow(gray,cmap='gray')
```

```python
!wget -c https://raw.githubusercontent.com/MarwaQabeel/Udacity-Computer-Vision-Nanodegree/master/Project%201-%20facial%20keypoint%20detection/detector_architectures/haarcascade_frontalface_default.xml
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

#  4 = scale factor. scale factor << means more faces
# 6 = minimum neighbours. min neigh << means more faces
faces = face_cascade.detectMultiScale(gray, 4, 6)
```

```python
# print out the detections found
print ('We found ' + str(len(faces)) + ' faces in this image')
print ("Their coordinates and lengths/widths are as follows")
print ('=============================')
print (faces)
```

```python
new_img = np.copy(image)

for(x,y,w,h) in faces:
    cv2.rectangle(new_img,(x,y),(x+w,y+w),(255,0,0),5)

plt.figure(figsize = (20,10))
plt.imshow(new_img)
```
